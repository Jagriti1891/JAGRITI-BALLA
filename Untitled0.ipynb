```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "day": [1, 2, 3, 4, 5],
    "delay": [10, 5, 15, 20, 8]
})
```

```python

data.dropna(inplace=True)
data['delay'] = data['delay'].astype(int)
```

```python
average_delay = data['delay'].mean()
std_delay = data['delay'].std()

print("Average Delay:", average_delay)
print("Standard Deviation:", std_delay)
```

```python
plt.figure()
plt.plot(data['delay'])
plt.xlabel("Index")
plt.ylabel("Delay")
plt.title("Delay Analysis")
plt.show()
```
