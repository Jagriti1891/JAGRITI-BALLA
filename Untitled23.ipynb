```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate hourly data for 30 days
date_range = pd.date_range(start="2025-01-01", periods=30*24, freq='H')

np.random.seed(42)
load = np.random.normal(loc=50, scale=10, size=len(date_range))

data = pd.DataFrame({
    "Datetime": date_range,
    "Load_kW": load
})

data.set_index("Datetime", inplace=True)
```

```python
data["Hour"] = data.index.hour

peak = data[(data["Hour"] >= 18) & (data["Hour"] <= 22)]["Load_kW"].mean()
off_peak = data[(data["Hour"] < 6)]["Load_kW"].mean()

print("Peak Load:", round(peak,2))
print("Off-Peak Load:", round(off_peak,2))
```

```python
data["Rolling_Avg"] = data["Load_kW"].rolling(window=24).mean()

plt.figure()
plt.plot(data["Load_kW"], label="Hourly Load")
plt.plot(data["Rolling_Avg"], label="24H Rolling Avg")
plt.legend()
plt.title("Load Trend Analysis")
plt.xlabel("Time")
plt.ylabel("Load (kW)")
plt.show()
```

```python
data["DayType"] = np.where(data.index.dayofweek < 5, "Weekday", "Weekend")

comparison = data.groupby("DayType")["Load_kW"].mean()
comparison.plot(kind="bar", title="Weekday vs Weekend Consumption")
plt.ylabel("Average Load (kW)")
plt.show()
```

```python
mean = data["Load_kW"].mean()
std = data["Load_kW"].std()

data["Anomaly"] = data["Load_kW"] > (mean + 2*std)

anomalies = data[data["Anomaly"] == True]
print("Anomalies Detected:", len(anomalies))
```

```python
efficiency_score = (off_peak / peak) * 100
print("Energy Efficiency Score:", round(efficiency_score,2))
```

```python
pivot = data.pivot_table(values="Load_kW",
                         index=data.index.hour,
                         columns=data.index.day,
                         aggfunc="mean")

plt.imshow(pivot, aspect='auto')
plt.colorbar(label="Load (kW)")
plt.title("Hourly Load Heatmap")
plt.xlabel("Day")
plt.ylabel("Hour")
plt.show()
```
